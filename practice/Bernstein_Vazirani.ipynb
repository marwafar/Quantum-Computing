```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secret_number='10100110'
n=len(secret_number)
```

```python
circuit=QuantumCircuit(n+1,n)
```

```python
circuit.h(range(n))
circuit.x(n)
circuit.h(n)
circuit.draw()
```

```python
circuit.barrier()
for index,char in enumerate(reversed(secret_number)):
    if char=='1':
        circuit.cx(index,n)
circuit.barrier()
circuit.draw()
```

```python
circuit.h(range(n))
circuit.draw()
```

```python
circuit.barrier()
circuit.measure(range(n),range(n))
circuit.draw()
```

```python
simulator= Aer.get_backend('qasm_simulator')
job=execute(circuit,backend=simulator,shots=1)
output=job.result()
counts=output.get_counts()
print(counts)
```
